##Task 2: Data Preparation and Preprocessing

>**Use sklearn.model_selection.train_test_split to create training and test sets**:

>>**For the Titanic dataset, use stratified sampling based on the Survived column. Justify your choice**

The survived column is the most important column as that is what we will be predicting. The data can be divided to properly represent this key chracteristic. The column is imbalanced and random sampling might accidentally skew this ratio, harming model performance. Stratified sampling preserves the class distribution, making the training set representative and avoiding bias.

```python
titanic["Survived"].hist()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(titanic, titanic["Survived"]):
    titanic_train = titanic.loc[train_index]
    titanic_test = titanic.loc[test_index]

# 712 training + 179 testing
#displaying proportion to double check
```

```python
titanic["Survived"].value_counts() / len(titanic)
```

```python
titanic_train["Survived"].value_counts() / len(titanic_train)
```

```python
titanic_test["Survived"].value_counts() / len(titanic_test)
```

>>**For the House Prices dataset, use random sampling. Justify your choice**

There are no inherent categories so we will go with the simplest and most straightforward type of sampling. This method will work well as our data is large enough and is relatively homogeneous i.e. there is no significant imbalance of features. It creates a reasonable representation of the whole dataset for both training and testing.

```python
from sklearn.model_selection import train_test_split

#80% training and 20% testing i.e. 1168 training + 292 testing
housing_train, housing_test = train_test_split(housing, test_size=0.2, random_state=42)
```

>**Split the data into features (X) and target (y) or labels for both training and testing sets**

```python
#creating function that wil perform the split
def split_data(target_dataset, target_label):
  return target_dataset.drop(target_label, axis=1), target_dataset[target_label].copy()

#storing in temporary variable as we want to make a copy of the raw data later on
titanic_train_X_temp, titanic_train_y = split_data(titanic_train, "Survived")
titanic_test_X_temp, titanic_test_y = split_data(titanic_test, "Survived")

housing_train_X, housing_train_y = split_data(housing_train, "SalePrice")
housing_test_X, housing_test_y = split_data(housing_test, "SalePrice")
```

##Task 3: Building a Processing Pipeline

>**Create a scikit-learn pipeline**: Imputation | Encoding | Scaling | Feature selection

```python
from sklearn.compose import ColumnTransformer

#Imputation of missing values: median for numeric, mode for categorical, complete removal for excessive missingness or irrelevance
from sklearn.impute import SimpleImputer
num_imputer = SimpleImputer(strategy="median")
cat_imputer = SimpleImputer(strategy="mode")
# Cabin attribute has too many missing values (almost 77%) which would render imputation ineffective so complete removal
titanic_train_X, titanic_test_X = titanic_train_X_temp.drop("Cabin", axis=1), titanic_test_X_temp.drop("Cabin", axis=1)
#Name attribute deosnt seem to be significant or relevant to predictions so complete removal
titanic_train_X, titanic_test_X = titanic_train_X.drop("Name", axis=1), titanic_test_X.drop("Name", axis=1)

#Encoding of categorical variables using OneHotEncoder
from sklearn.preprocessing import OneHotEncoder

#Scaling of numeric features using StandardScaler
from sklearn.preprocessing import StandardScaler

#Feature selection using SelectKBest with f_regression/f_classif
from sklearn.feature_selection import SelectKBest, f_classif, f_regression
```

```python
from sklearn.pipeline import Pipeline

#Numerical Pipeline
num_pipeline = Pipeline([
    ('imputer', num_imputer),
    ('std_scaler', StandardScaler()),
])

#Categorical Pipeline
cat_pipeline = Pipeline([
    ('cat_encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False)),
    #handle_unknown='ignore' ignores unseen categories in test data instead of crashing
])
```

>>**Titanic Pipeline**

```python
#split into numerical and categorical values
titanic_cat_attr = list(titanic_train_X.select_dtypes(include=['object']))
titanic_num_attr = list(titanic_train_X.select_dtypes(include=['int64', 'float64']))

#ColumnTransformer applies different transformations to different columns and them concatenated
titanic_preprocessor = ColumnTransformer([
    ("num", num_pipeline, titanic_num_attr),
    ("cat", cat_pipeline, titanic_cat_attr)
])

#full transformation pipeline
titanic_pipeline = Pipeline([
    ('preprocessor', titanic_preprocessor),
    ('selector', SelectKBest(score_func=f_classif, k=10))
])

#transforming our data
titanic_train_X_prepared = titanic_pipeline.fit_transform(titanic_train_X, titanic_train_y)

#Logistic Regression Model
from sklearn.linear_model import LogisticRegression

#fitting our model
titanic_log_reg = LogisticRegression()
titanic_log_reg.fit(titanic_train_X_prepared, titanic_train_y)
```

>>**Housing Pipeline**

```python
#split into numerical and categorical values
housing_cat_attr = list(housing_train_X.select_dtypes(include=['object']))
housing_num_attr = list(housing_train_X.select_dtypes(include=['int64', 'float64']))

#ColumnTransformer applies different transformations to different columns and them concatenated
housing_preprocessor = ColumnTransformer([
    ("num", num_pipeline, housing_num_attr),
    ("cat", cat_pipeline, housing_cat_attr)
])

#full transformation pipeline
housing_pipeline = Pipeline([
    ('preprocessor', housing_preprocessor),
    ('selector', SelectKBest(score_func=f_regression, k=35))
])

#transforming our data
housing_train_X_prepared = housing_pipeline.fit_transform(housing_train_X, housing_train_y)

#Linear Regression Model
from sklearn.linear_model import LinearRegression

#fitting our model
housing_lin_reg = LinearRegression()
housing_lin_reg.fit(housing_train_X_prepared, housing_train_y)
```

##Task 4: Model Evaluation and Comparison

>**Predicting from the testing dataset**

```python
#Titanic Predictions
titanic_test_X_prepared = titanic_pipeline.transform(titanic_test_X)
titanic_predictions = titanic_log_reg.predict(titanic_test_X_prepared)

#Housing Predictions
housing_test_X_prepared = housing_pipeline.transform(housing_test_X)
housing_predictions = housing_lin_reg.predict(housing_test_X_prepared)
```

>**Evaluate the model using appropriate metrics**:  

>>Titanic Classification Metrics: **Accuracy, Precision, Recall, F1-score**

```python
#Good classification metric is often above 70%, potentially reaching 95%+

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

#function for calculating different classification metrics
def classification_metrics(actual_y, predictions):
  print("Accuracy: ", (accuracy_score(actual_y, predictions)))
  print("Precision: ", (precision_score(actual_y, predictions, zero_division=0)))
  print("Recall: ", (recall_score(actual_y, predictions, zero_division=0)))
  print("F1-score: ", (f1_score(actual_y, predictions, zero_division=0)))

classification_metrics(titanic_test_y, titanic_predictions)
```

>>Housing Regression Metric: **RMSE and RÂ² score**  

```python
#RMSE: the lower, the better a given model is able to â€œfitâ€ a dataset
from sklearn.metrics import mean_squared_error
#RÂ² score: Best possible score is 1.0
from sklearn.metrics import r2_score

#function for calculating different regression metrics
def regression_metrics(actual_y, predictions):
  print("RMSE:", (np.sqrt(mean_squared_error(actual_y, predictions))))
  print("RÂ² score:", (r2_score(actual_y, predictions)))

regression_metrics(housing_test_y, housing_predictions)
```

Our models do not seem to have very large errors, but there's always room for improvement!

>**Compare performance with a baseline model (Majority class for Classification, Mean prediction for Regression)**

A baseline model is a simple model used to predict the outcome of data. It serves as a reference point in machine learning tasks and allows us to assess the performance of more complex models. If a model fails to outperform the baseline, it suggests issues with the approach or data.

A dummy is a simple machine learning model that predicts numerical values using basic rules, without actually learning from the input data. It serves as a baseline for comparing the performance of more complex models.

**Majority Class Model**: where the most frequent class in the data is predicted for all observations

**Mean Prediction Model**: average of the target variable is predicted for all instances

```python
# Titanic | Classification | Majority Class

from sklearn.dummy import DummyClassifier

#creating and training dummy model then predicting
titanic_base_model = DummyClassifier(strategy="most_frequent")
titanic_base_model.fit(titanic_train_X_prepared, titanic_train_y)
titanic_base_predictions = titanic_base_model.predict(titanic_test_X_prepared)

# Comparing metrics of both models
print("Baseline Model Accuracy:", (classification_metrics(titanic_test_y, titanic_base_predictions)))
print()
print("Our Model Accuracy:", (classification_metrics(titanic_test_y, titanic_predictions)))
```

```python
# Housing | Regression | Mean Prediction

from sklearn.dummy import DummyRegressor

#creating and training dummy model then predicting
housing_base_model = DummyRegressor(strategy="mean")
housing_base_model.fit(housing_train_X_prepared, housing_train_y)
housing_base_predictions = housing_base_model.predict(housing_test_X_prepared)

# Comparing metrics of both models
print("Baseline Model RMSE:")
regression_metrics(housing_test_y, housing_base_predictions)
print()
print("Our Model RMSE:")
regression_metrics(housing_test_y, housing_predictions)
```

Our models are doing significantly better than the baseline dummy models!

##Task 5: Feature Engineering

>**Create at least two new features for each dataset:**

>>Titanic: **FamilySize, Title extracted from Name**

```python
#copying unmanipulated data into new variables
titanic_train_X_new = titanic_train_X_temp.copy()
titanic_test_X_new = titanic_test_X_temp.copy()

#Creating new variable FamilySize = parents + children + siblings + spouse
titanic_train_X_new["FamilySize"] = titanic_train_X_new["Parch"] + titanic_train_X_new["SibSp"]
titanic_test_X_new["FamilySize"] = titanic_test_X_new["Parch"] + titanic_test_X_new["SibSp"]

#Creating new variable Title that will be extracted from Name
titanic_train_X_new["Title"] = titanic_train_X_new["Name"].str.extract(r' ([A-Za-z]+)\.', expand=False)
titanic_test_X_new["Title"] = titanic_test_X_new["Name"].str.extract(r' ([A-Za-z]+)\.', expand=False)
#grouping rare titles together
# Group rare titles
titanic_train_X_new["Title"] = titanic_train_X_new["Title"].replace(["Lady", "Countess","Capt", "Col","Don", "Dr", "Major", "Rev", "Sir", "Jonkheer", "Dona"], "Rare")
titanic_test_X_new["Title"] = titanic_test_X_new["Title"].replace(["Lady", "Countess","Capt", "Col","Don", "Dr", "Major", "Rev", "Sir", "Jonkheer", "Dona"], "Rare")
```

>>Housing: **AgeOfHouse, TotalSquareFootage**

```python
#copying unmanipulated data into new variables
housing_train_X_new = housing_train_X.copy()
housing_test_X_new = housing_test_X.copy()

#Creating new variable AgeOfHouse = current year - year it was built
housing_train_X_new["AgeOfHouse"] = 2026 - housing_train_X_new["YearBuilt"]

#Creating new variable TotalSquareFootage = Sum of Square footages across all floors
housing_train_X_new["TotalSquareFootage"] = housing_train_X_new["TotalBsmtSF"]+housing_train_X_new["1stFlrSF"]+housing_train_X_new["2ndFlrSF"]
```

>**Retrain your pipeline with the new features and evaluate if performance improves**

```python
#Titanic Dataset

#transforming training data
titanic_train_X_new_prepared = titanic_pipeline.fit_transform(titanic_train_X_new, titanic_train_y)

#creating and training model on new data
titanic_new_log_reg = LogisticRegression()
titanic_new_log_reg.fit(titanic_train_X_new_prepared, titanic_train_y)

#trasforming testing data and predicting
titanic_test_X_new_prepared = titanic_pipeline.transform(titanic_test_X_new)
titanic_new_predictions = titanic_new_log_reg.predict(titanic_test_X_new_prepared)

#re-evaluating metrics
classification_metrics(titanic_test_y, titanic_new_predictions)
```

```python
#Housing Dataset

#transforming training data
housing_train_X_new_prepared = housing_pipeline.fit_transform(housing_train_X_new, housing_train_y)

#creating and training model on new data
housing_new_lin_reg = LinearRegression()
housing_new_lin_reg.fit(housing_train_X_new_prepared, housing_train_y)

#trasforming testing data and predicting
housing_test_X_new_prepared = housing_pipeline.transform(housing_test_X_new)
housing_new_predictions = housing_new_lin_reg.predict(housing_test_X_new_prepared)

#re-evaluating metrics
regression_metrics(housing_test_y, housing_new_predictions)
```

There was no change in the performance, meaning the new features we added did not have any predictive value. Feature engineering doesn't always improve accuracy and this is completely normal.
